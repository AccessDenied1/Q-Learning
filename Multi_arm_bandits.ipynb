```
import numpy as np
class MAB:
    def Arm1(self):
        arm1=np.random.uniform(0.7,0.4)
        return arm1
    def Arm2(self):
        arm2=np.random.uniform(0.6,0.5)
        return arm2
    def UCB(self):
        arm1=self.Arm1()
        arm2=self.Arm2()
        S1=arm1
        S2=arm2
        N1=1.0
        N2=1.0
        t=0
        for t in range(1,100):
            sample_mean1=S1/N1
            sample_mean2=S2/N2
            confidence_value1=np.sqrt((2*np.log(t))/N1)
            confidence_value2=np.sqrt((2*np.log(t))/N2)
            UCB1=sample_mean1+confidence_value1
            UCB2=sample_mean2+confidence_value2
            action=np.argmax([UCB1,UCB2])
            if action==0:
                arm1=self.Arm1()
                S1+=arm1
                N1+=1.0
            else:
                arm2=self.Arm2()
                S2+=arm2
                N2+=1.0
            print("Selected arm is,",action+1)
        print('sample_mean1',sample_mean1)
        print('sample_mean2',sample_mean2)
obj=MAB()
obj.UCB()
```
